# Fatal Police Shootings Data

The Washington Post's database contains records of every fatal shooting in the United States by a police officer in the line of duty since Jan. 1, 2015.

https://github.com/washingtonpost/data-police-shootings

```python
# Import dependencies
%matplotlib inline
from matplotlib import style
style.use('fivethirtyeight')
import matplotlib.pyplot as plt
import matplotlib.font_manager
from matplotlib import rc
import matplotlib.gridspec as gridspec
import scipy.stats as stats
import numpy as np
import pandas as pd
import datetime as dt
from datetime import timedelta, date
import seaborn as sns
```

```python
# Read CSV File
df = pd.read_csv("/Users/nicolemuscanell/UCF-11-HTML-CSS/Data/shootings.csv")
df.head()
```

```python
# Creating DF
shootings_df = pd.DataFrame(df, columns=['id', 'name', 'date', 'manner_of_death', 'armed', 'age', 'gender', 'race', 
                                         'city', 'state', 'signs_of_mental_illness', 'threat_level', 'flee',
                                         'body_camera', 'arms_category'])
shootings_df.head()
```

```python
# Counts for shootings by race
race_counts = shootings_df.groupby(['race'])["id"].count().sort_values(ascending=False)
#race_counts

# Total n size
total_n = shootings_df.groupby(['race'])["id"].count().sum()
#total_n

# Percentages for shootings by race
race_percent = (race_counts/total_n)*100
race_percent
```

```python
# Plot for race

csfont = {'fontname': "Georgia"}

race_percent.plot.bar(x='race', y='percent', rot=70, legend=False, color='deeppink', figsize=(10,5))
plt.title('Fatal Police Shootings: Percentages by Race', fontsize=14, fontweight='bold', **csfont)
plt.ylabel("Fatal Shootings (Percent)", fontsize=14, fontweight='bold', **csfont)
plt.xlabel("Victim Race", fontsize=14, fontweight='bold', **csfont)
plt.xticks(fontsize=14, fontweight='bold', **csfont)
plt.ylim(0,100,10)
plt.tight_layout()
```
